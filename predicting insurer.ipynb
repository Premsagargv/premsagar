# Predicting the  Degree of Regulatory Compliance  of the Insurers  through  evaluation of complaints resolution system  & Pattern Extraction

# Problem Statement

##### Some Insurers pose substantial threats to customers, yet early identification of them is notoriously difficult. We aimed to develop an algorithm for use by regulators Compliance in prospectively identifying Customers  of attracting formal complaints about Insurance,through  evaluation of complaints .

# Business Understanding

#### The performance of Insurance Companies depends on the work they do daily.Close review of available information on evaluation of complaints may help to identify troubling patterns and imminent risks to customers.

```python
df5 = df4.drop(['RecoveredFromInsurer','Days'],axis=True) 
df5.head()
```

```python
df5.shape
```

```python
x = df5.copy().drop('DRC',axis = 1)
y = df5['DRC']
```

```python
from sklearn.preprocessing import Imputer
mean_imputer = Imputer()
x=pd.DataFrame(mean_imputer.fit_transform(x),columns = x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.30)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr_model = LogisticRegression()
lr_model.fit(x_train,y_train)
pred_train = lr_model.predict(x_train)
pred_test = lr_model.predict(x_test)
print("Accuracy on train is:",accuracy_score(y_train,pred_train))
print("Accuracy on test is:",accuracy_score(y_test,pred_test))
```

```python
## Decision Tree
from sklearn import tree
dtc = tree.DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_pred = dtc.predict(x_test)
#pred_test = dtc.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
print("Accuracy on train is:",dtc.score(x_train,y_train))
print("Accuracy on test is:",dtc.score(x_test,y_test))
```

```python
## Random Forest
from sklearn.ensemble import RandomForestClassifier

rfc_model = RandomForestClassifier(n_estimators = 100,random_state = 99)
rfc_model.fit(x_train,y_train.values.ravel())
predicted = rfc_model.predict(x_test)
#print('Random Forest',accuracy_score(y_test,predicted))
pred_train = rfc_model.predict(x_train)
pred_test = rfc_model.predict(x_test)
print("Accuracy on train is:",accuracy_score(y_train,pred_train))
print("Accuracy on test is:",accuracy_score(y_test,pred_test))
```
